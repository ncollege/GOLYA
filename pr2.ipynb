```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")

print("First 5 Rows of Dataset")
print(df.head())
```

```python
print("\nDataset Information")
print(df.info())

print("\nShape of Dataset")
print(df.shape)
```

```python
print("\nMissing Values in Dataset")
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df['Cabin'] = df['Cabin'].fillna('Unknown')

print("\nMissing Values After Handling")
print(df.isnull().sum())
```

```python
print("\nUnique Values in Sex Column")
print(df['Sex'].unique())

print("\nUnique Values in Embarked Column")
print(df['Embarked'].unique())
```

```python
df['Sex'] = df['Sex'].str.lower()

print("\nUpdated Unique Values in Sex Column")
print(df['Sex'].unique())
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("\nLower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

outliers = df[(df['Fare'] < lower_limit) | (df['Fare'] > upper_limit)]

print("\nOutliers in Fare Column")
print(outliers[['Fare']])

print("\nNumber of Outliers:", len(outliers))
```

```python
df_no_outliers = df[
    (df['Fare'] >= lower_limit) &
    (df['Fare'] <= upper_limit)
]

print("\nShape Before Removing Outliers")
print(df.shape)

print("\nShape After Removing Outliers")
print(df_no_outliers.shape)
```

```python
df_no_outliers['Fare_Log'] = np.log1p(df_no_outliers['Fare'])

print("\nOriginal Fare Values")
print(df_no_outliers['Fare'].head())

print("\nLog Transformed Fare Values")
print(df_no_outliers['Fare_Log'].head())
```

```python
plt.figure(figsize=(8,5))

plt.hist(df_no_outliers['Fare'], bins=30)

plt.title("Fare Distribution Before Transformation")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df_no_outliers['Fare_Log'], bins=30)

plt.title("Fare Distribution After Log Transformation")
plt.xlabel("Fare Log")
plt.ylabel("Frequency")

plt.show()
```

```python
print("\nFinal Dataset")
print(df_no_outliers.head())

print("\nFinal Dataset Information")
print(df_no_outliers.info())
```
