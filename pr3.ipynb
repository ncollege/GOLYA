```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Iris.csv")

print("First 5 Rows of Dataset")
print(df.head())
```

```python
print("\nDataset Information")
print(df.info())

print("\nShape of Dataset")
print(df.shape)

print("\nColumn Names")
print(df.columns)
```

```python
print("\nSummary Statistics of SepalLengthCm grouped by Species")

grouped_data = df.groupby('Species')['SepalLengthCm']

print(grouped_data.describe())
```

```python
setosa = df[df['Species'] == 'Iris-setosa']['SepalLengthCm'].tolist()

versicolor = df[df['Species'] == 'Iris-versicolor']['SepalLengthCm'].tolist()

virginica = df[df['Species'] == 'Iris-virginica']['SepalLengthCm'].tolist()

print("\nList of Sepal Length for Iris-setosa")
print(setosa)

print("\nList of Sepal Length for Iris-versicolor")
print(versicolor)

print("\nList of Sepal Length for Iris-virginica")
print(virginica)
```

```python
def statistical_details(species_name):

    species_data = df[df['Species'] == species_name]

    print("\n===================================")
    print("Species :", species_name)
    print("===================================")

    print("\nMean")
    print(species_data.mean(numeric_only=True))

    print("\nMedian")
    print(species_data.median(numeric_only=True))

    print("\nMinimum Values")
    print(species_data.min(numeric_only=True))

    print("\nMaximum Values")
    print(species_data.max(numeric_only=True))

    print("\nStandard Deviation")
    print(species_data.std(numeric_only=True))

    print("\nPercentiles")
    print(species_data.quantile([0.25, 0.50, 0.75], numeric_only=True))
    
statistical_details('Iris-setosa')

statistical_details('Iris-versicolor')

statistical_details('Iris-virginica')
```
