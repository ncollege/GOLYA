```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
document = """
Data Analytics is the process of analyzing data
to get useful information. Machine Learning and
Artificial Intelligence are important technologies.
"""

print("Original Document:\n")
print(document)
```

```python
tokens = word_tokenize(document)

print("\nTokens:\n")
print(tokens)
```

```python
pos = pos_tag(tokens)

print("\nPOS Tagging:\n")
print(pos)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = []

for word in tokens:

    if word.lower() not in stop_words:

        filtered_words.append(word)

print("\nAfter Stop Word Removal:\n")
print(filtered_words)
```

```python
ps = PorterStemmer()

stemmed_words = []

for word in filtered_words:

    stemmed_words.append(ps.stem(word))

print("\nStemmed Words:\n")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = []

for word in filtered_words:

    lemmatized_words.append(lemmatizer.lemmatize(word))

print("\nLemmatized Words:\n")
print(lemmatized_words)
```

```python
documents = [document]

vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(documents)

print("\nTF-IDF Feature Names:\n")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Values:\n")
print(tfidf.toarray())
```
